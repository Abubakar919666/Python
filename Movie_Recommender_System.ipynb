```python
# need to get the values of the name key in list
# but it is in form of a string
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# test case
convert(movies.iloc[0].genres)
```

```python
movies.genres = movies.genres.apply(convert)
movies.genres
```

```python
convert(movies.keywords)
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
# top 3 cast
movies.cast = movies.cast.apply(convert_cast)
movies.cast
```

```python
movies['crew'][0]
```

```python
def get_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies.crew = movies.crew.apply(get_director)
movies.crew
```

```python
movies.crew = movies['crew'].apply(get_director)
```
