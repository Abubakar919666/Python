```python
# Load and Display Data â†’ Read the dataset and print the first 5 rows.
data = pd.read_csv('covid_19_country_wise_latest.csv')
data.head()
```

```python
data.tail()
```

```python
# Set the Dataset index is Unnamed: 0
data.set_index("Unnamed: 0", inplace=True)
data.head()
```

```python
# Now replace the index column Unnamed: 0 name with index
data.rename_axis("index", inplace=True)
data.head()
```

```python
# Function to replace whitespace in column names with underscores
def replace_whitespace(Data):

    # Replace whitespaces with underscores in column names
    data.columns = data.columns.str.replace(' ', '_')

    # Return the updated DataFrame
    return data


data = replace_whitespace(data)

data.head()
```

```python
column_names = data.columns
print(column_names)
```

```python
data_types = data.dtypes
print(data_types)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# fill the nan values with the mean of the column
data.fillna(data.mean())
```

```python
# Fill NaN values with the mean for numeric columns only
data[data.select_dtypes(include=['number']).columns] = data.select_dtypes(include=['number']).fillna(data.mean())

# Check the updated dataset
print(data)
```

```python
# Select only numeric columns
numeric_columns = data.select_dtypes(include=['number']).columns

# Fill NaN values in numeric columns with the mean
data.fillna(data[numeric_columns].mean())

# Check the updated dataset
print(data)
```
